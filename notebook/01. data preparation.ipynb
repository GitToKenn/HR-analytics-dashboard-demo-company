## Education

```python
df_raw["Education"].value_counts()
```

```python
edu_rules = {
    "Sales Representative": ["High School","Associates Degree","Bachelor's Degree"],
    "Laboratory Technician": ["Associates Degree","Bachelor's Degree","Master's Degree"],
    "Human Resources": ["Bachelor's Degree","Master's Degree","Doctoral Degree"],
    "Research Scientist": ["Bachelor's Degree","Master's Degree","Doctoral Degree"],
    "Sales Executive": ["Bachelor's Degree","Master's Degree","Doctoral Degree"],
    "Healthcare Representative": ["Bachelor's Degree","Master's Degree","Doctoral Degree"],
    "Manager": ["Bachelor's Degree","Master's Degree","Doctoral Degree"],
    "Research Director": ["Master's Degree","Doctoral Degree"],
    "Manufacturing Director": ["Bachelor's Degree","Master's Degree","Doctoral Degree"]
}

edu_probs = {
    "Sales Representative": [0.3, 0.4, 0.3],
    "Laboratory Technician": [0.3, 0.5, 0.2],
    "Human Resources": [0.6, 0.3, 0.1],
    "Research Scientist": [0.3, 0.5, 0.2],
    "Sales Executive": [0.5, 0.4, 0.1],
    "Healthcare Representative": [0.4, 0.4, 0.2],
    "Manager": [0.3, 0.5, 0.2],
    "Research Director": [0.4, 0.6],
    "Manufacturing Director": [0.3, 0.5, 0.2]
}

def assign_education(role):
    if role in edu_rules:
        return np.random.choice(edu_rules[role], p=edu_probs[role])
    else:
        # fallback in case a role was missed
        return np.random.choice(["Bachelor's Degree","Master's Degree"], p=[0.7,0.3])

df_bucket["Education"] = df_bucket["Job Role"].apply(assign_education)
```

```python
#rebalance

df_bucket["Education"] = df_bucket["Job Role"].apply(assign_education)

from collections import Counter

edu_targets = {
    "High School": 170,
    "Associates Degree": 282,
    "Bachelor's Degree": 572,
    "Master's Degree": 398,
    "Doctoral Degree": 48
}

# current counts
edu_current = df_bucket["Education"].value_counts().to_dict()
diffs = {k: edu_current.get(k, 0) - v for k, v in edu_targets.items()}

# build deficit pool
deficits = {k: -d for k, d in diffs.items() if d < 0}
deficit_pool = Counter(deficits)

# rebalance surplus → into deficits
for degree, surplus in diffs.items():
    if surplus > 0:  # too many of this degree
        candidates = df_bucket[df_bucket["Education"] == degree]
        n_reassign = min(len(candidates), surplus, sum(deficit_pool.values()))
        if n_reassign > 0:
            idx = candidates.sample(n_reassign, random_state=42).index
            deficit_items = list(deficit_pool.elements())
            new_values = np.random.choice(deficit_items, size=n_reassign, replace=True)
            df_bucket.loc[idx, "Education"] = new_values
            for v in new_values:
                if deficit_pool[v] > 0:
                    deficit_pool[v] -= 1
```

## Years At Company

```python
print(df_raw["Years At Company"].value_counts())
```

```python
def assign_years_at_company(row):
    twy = int(row["Total Working Years"])
    if twy <= 0:
        return 0

    lvl = int(row["Job Level"])

    # Define job-level preferred bins
    if lvl == 1:
        pref_bins  = list(range(0, 6))   # 0–5
        pref_probs = [0.25, 0.25, 0.20, 0.15, 0.10, 0.05]
    elif lvl == 2:
        pref_bins  = list(range(1, 9))   # 1–8
        pref_probs = [0.20,0.20,0.18,0.15,0.12,0.08,0.05,0.02]
    elif lvl == 3:
        pref_bins  = list(range(2, 13))  # 2–12
        pref_probs = [0.15,0.15,0.15,0.12,0.10,0.08,0.07,0.06,0.05,0.04,0.02]
    elif lvl == 4:
        pref_bins  = list(range(3, 16))  # 3–15
        pref_probs = [0.12,0.12,0.12,0.11,0.10,0.09,0.08,0.07,0.06,0.05,0.04,0.02,0.02]
    else:  # lvl == 5
        pref_bins  = list(range(5, 21))  # 5–20
        pref_probs = [0.10,0.10,0.10,0.10,0.09,0.09,0.08,0.08,0.07,0.05,
                      0.05,0.04,0.03,0.02,0.01,0.01]

    # Trim bins to available TWY
    bins = [b for b in pref_bins if b <= twy]

    # Fallback: if bins became empty, just use 0..twy
    if not bins:
        bins = list(range(0, twy+1))
        probs = np.ones(len(bins)) / len(bins)
    else:
        probs = np.array(pref_probs[:len(bins)])
        probs = probs / probs.sum()  # normalize

    return int(np.random.choice(bins, p=probs))

df_bucket["Years At Company"] = df_bucket.apply(assign_years_at_company, axis=1)
```

```python
def assign_years_in_role(row):
    # cap: cannot exceed Years At Company
    cap = row["Years At Company"]
    if cap <= 0:
        return 0
    
    # biased brackets (probabilities add to 1)
    brackets = [
        (0, 2, 0.10),   # 0–1 years
        (2, 7, 0.60),   # 2–6 years
        (7, 11, 0.20),  # 7–10 years
        (11, cap+1, 0.10)  # 11+ years (up to cap)
    ]
    
    # pick a bracket
    choices = [b for b in brackets if b[0] < cap+1]  # only brackets ≤ cap
    probs = [p for (_, _, p) in choices]
    probs = [p/sum(probs) for p in probs]  # re-normalize
    
    low, high, _ = choices[np.random.choice(len(choices), p=probs)]
    
    # generate random years inside bracket, but not above cap
    return np.random.randint(low, min(high, cap+1))

df_bucket["Years In Current Role"] = df_bucket.apply(assign_years_in_role, axis=1)
```

## Years Since Last Promotion

```python
print(df_raw["Years Since Last Promotion"].value_counts())
```

```python
def assign_years_since_promo(row):
    cap = row["Years At Company"]
    if cap <= 0:
        return 0
    
    # Slightly looser on 10+ years
    brackets = [
        (0, 3, 0.48),      # 0–2 years (dominant, but trimmed a bit)
        (3, 6, 0.25),      # 3–5 years
        (6, 11, 0.20),     # 6–10 years
        (11, cap+1, 0.07)  # 11+ years (boosted from 0.05 → 0.07)
    ]
    
    # Only keep brackets possible within employee tenure
    choices = [b for b in brackets if b[0] < cap+1]
    probs = [p for (_, _, p) in choices]
    probs = [p/sum(probs) for p in probs]  # normalize
    
    low, high, _ = choices[np.random.choice(len(choices), p=probs)]
    
    return np.random.randint(low, min(high, cap+1))

df_bucket["Years Since Last Promotion"] = df_bucket.apply(assign_years_since_promo, axis=1)
```

## Years With Curr Manager

```python
print(df_raw["Years With Curr Manager"].value_counts())
```

```python
def assign_years_with_manager(job_level):
    if job_level == 1:  # entry
        bins = [0, 1, 2, 3, 4, 5]
        probs = [0.40, 0.25, 0.15, 0.10, 0.06, 0.04]
    elif job_level == 2:  # early-mid
        bins = [1, 2, 3, 4, 5, 6, 7, 8]
        probs = [0.25, 0.25, 0.20, 0.15, 0.08, 0.04, 0.02, 0.01]
    elif job_level == 3:  # mid
        bins = [2, 3, 4, 5, 6, 7, 8, 9, 10]
        probs = [0.20, 0.20, 0.15, 0.15, 0.10, 0.08, 0.06, 0.04, 0.02]
    else:  # senior 4–5
        bins = list(range(3, 16))  # 3–15
        probs = [0.15, 0.15, 0.15, 0.12, 0.10, 0.08, 0.07, 0.05, 0.05, 0.04, 
                 0.02, 0.01, 0.01]
    
    probs = np.array(probs) / np.sum(probs)
    return np.random.choice(bins, p=probs)

df_bucket["Years With Curr Manager"] = df_bucket["Job Level"].apply(assign_years_with_manager)
```
